### manually set the layout of the interface using the `layout` parameter. The layout can be set to either "horizontal" or "vertical", which determines how the input and output components are arranged in the interface. This allows for customization of the user interface to better suit the needs of the application.

```python
with gr.Blocks() as demo:
    
    with gr.Row():
        # here goes the input components inside the row
        text1=gr.Text(value="Text 1")
        text2=gr.Text(value="Text 2")
        
    with gr.Row():
        # here goes the output components inside the 2nd row
        text3=gr.Text(value="Text 3")
        text4=gr.Text(value="Text 4")

demo.launch()
```

```python
with gr.Blocks() as demo:
    with gr.Row():
        with gr.Column():
            text1=gr.Text(value="row 1, column 1 component 1")
            text2=gr.Text(value="row 1, column 1 component 2")
        with gr.Column():
            text3=gr.Text(value="row 1, column 2 ")
    with gr.Row():
        text4=gr.Text(value="row 2, column 1")
        
demo.launch()
```

```python
with gr.Blocks() as demo:
    with gr.Row():
        with gr.Column(scale=2): # occupy 2/3 of the row
            text1=gr.Text(value="row 1, column 1 component 1")
            text2=gr.Text(value="row 1, column 1 component 2")
        with gr.Column(scale=1): # occupy 1/3 of the row
            text3=gr.Text(value="row 1, column 2 ")
    with gr.Row():
        text4=gr.Text(value="Botton row")
        
demo.launch()
```

# 2 types of Events in gradio
#### change event -> trigerred when one component value is change 
#### click event -> trigerred when one component is clicked

### following are change event example

```python
def multiply(num1,num2):
    return num1*num2

with gr.Blocks() as demo:
    with gr.Row():
        x_slider=gr.Slider()
        y_slider=gr.Slider()
    with gr.Row():
        result=gr.Text()
    # when   
    y_slider.change(fn=multiply,inputs=[x_slider,y_slider],outputs=result)

demo.launch()
```

```python
with gr.Blocks() as demo:
    with gr.Row():
        x_slider=gr.Slider()
        y_slider=gr.Slider()
    with gr.Row():
        result=gr.Text()
    # when   
    y_slider.change(fn=multiply,inputs=[x_slider,y_slider],outputs=result)
    x_slider.change(fn=multiply,inputs=[x_slider,y_slider],outputs=result)
    
demo.launch()
```

### click event example -> button click trigger event

```python
with gr.Blocks() as demo:
    with gr.Row():
        x_slider=gr.Slider()
        y_slider=gr.Slider()
        
    with gr.Row():
        result=gr.Text()
        
    with gr.Row():
        button=gr.Button("Multiply !")
    
    button.click(
        fn=multiply,
        inputs=[x_slider,y_slider],
        outputs=result
    )
demo.launch()
```

### return different outputs

```python
def make_grayscale(image_path):
    # open the image
    image=Image.open(image_path)
    
    # convert the image to grayscale
    image_grayscale=image.convert("L")
    
    return image_grayscale, "Image Converted to Grayscale"

gray_image=make_grayscale("image.png")
```

```python
with gr.Blocks() as demo:
    with gr.Row():
        input_image=gr.Image(type="filepath")
        output_image=gr.Image()
        
    with gr.Row():
        log=gr.Text()
        submit=gr.Button(value="convert to grayscale") # cal to action
        
    submit.click(fn=make_grayscale,
                 inputs=input_image,
                 outputs=[output_image,log])

demo.launch()
```

### gradio intergration with machine learning 


```python
from transformers import AutoImageProcessor, AutoModelForImageClassification

processor= AutoImageProcessor.from_pretrained("microsoft/resnet-18")
model=AutoModelForImageClassification.from_pretrained("microsoft/resnet-18")
```

```python
model
```

```python
def classify_image(image):
    
    # processing image
    image=processor(image,return_tensors='pt')["pixel_values"]
    
    # passing the image to the model -> logits 0-1000
    logits=model(image).logits
    
    # grab the index of the highest probability
    predicted_label=logits.argmax(-1).item()
    
    return model.config.id2label[predicted_label]
```

```python
iface=gr.Interface(fn=classify_image,
                   inputs=gr.Image(),
                   outputs=gr.Label(),
                   description="upload an image")
iface.launch()
```

### gradio sentiment analysis

```python
from transformers import pipeline
sentiment_analysis=pipeline("sentiment-analysis")

```

```python
sentiment_analysis("hello my name is happy and i am feeling very bad and sad today")
```

```python
def predict_sentiment(text):
    result=sentiment_analysis(text)
    return result[0]["label"],result[0]["score"]
```

```python
predict_sentiment("hello my name is happy and i am feeling very bad and sad today")
```

```python
ifac=gr.Interface(
    fn=predict_sentiment,
    inputs=gr.Textbox(lines=2,placeholder="Type your text here: "),
    outputs=[gr.Text(label="Sentiment"),gr.Text(label="Score")]
)

ifac.launch()
```

## gradio errors


```python
def validate_even(num):
    return num%2==0

def add_even(num1,num2):
    if not validate_even(num1) and not validate_even(num2):
        raise gr.Error("Both numbers are not even")
    if not validate_even(num1):
        raise gr.Error("num1 is not even")
    if not validate_even(num2):
        raise gr.Error("num2 is not even")
    return num1 + num2

```

```python
iface=gr.Interface(fn=add_even,
                   inputs=[gr.Number(),gr.Number()],
                   outputs=gr.Number())
iface.launch()
```
